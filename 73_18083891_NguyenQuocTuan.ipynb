### Yêu cầu: Đọc dữ liệu giá bất động sản montreal dùng hàm read_csv của pandas

Tách dữ liệu ra làm x và y

```python
# YOUR CODE HERE
x = df_montreal.to_numpy()
x = x[:,14:-1]
print(f'x_shape = {x.shape}')


y = df_montreal.to_numpy()
y = x[:,-1:]
y = y.reshape(-1)
print(f'y_shape = {y.shape}')

# YOUR CODE HERE
```

```python
# Tách dữ liệu thành training và testing set
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 42)
```

### Huấn luyện thử một số mô hình để xem hiệu năng

```python
# YOUR CODE HERE
linear = LinearRegression()
linear_reg = linear.fit(x_train, y_train)
# YOUR CODE HERE
```

```python
print(f'Kết quả dự đoán: {linear_reg.predict(x_test[0:5])} ')
print(f'Kết quả thực tế: {y_test[0:5]} ')
```

```python
# Testing models
y_pred_train = linear_reg.predict(x_train)
y_pred = linear_reg.predict(x_test)
print(f"MSE on training: {mean_squared_error(y_train,y_pred_train):.3f}, on testing: {mean_squared_error(y_test, y_pred):.3f}")
print(f"MAE on training: {mean_absolute_error(y_train, y_pred_train):.3f}, on testing: {mean_absolute_error(y_test, y_pred):.3f}")
```

Linear Regression là một mô hình cơ bản, có thể sẽ không phù hợp trong nhiều trường hợp thực tế khi dữ liệu phức tạp hơn. Vì vậy, ta nên thử nghiệm một số mô hình khác

```python
from sklearn.linear_model import Ridge, Lasso
# YOUR CODE HERE
ridge = Ridge(alpha=0.1, normalize=True).fit(x_train, y_train)
print(f'Kết quả dự đoán: {ridge.predict(x_test[0:5])} ')
print(f'Kết quả thực tế: {y_test[0:5]} ')
# YOUR CODE HERE
```

```python
# Testing models
y_pred_train = ridge.predict(x_train)
y_pred = ridge.predict(x_test)

print(f"MSE on training: {mean_squared_error(y_train, y_pred_train):.3f}, on testing: {mean_squared_error(y_test, y_pred):.3f}")
print(f"MAE on training: {mean_absolute_error(y_train, y_pred_train):.3f}, on testing: {mean_absolute_error(y_test, y_pred):.3f}")
```

Yêu cầu: Sử dụng hàm cross_validate của sklearn để thực hiện cross_validation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate

# YOUR CODE HERE
estimator = LogisticRegression()
scoring = ['precision_macro', 'recall_macro']
scores = cross_validate(estimator, x, y_new, scoring=scoring, return_train_score=True, return_estimator=True)

# YOUR CODE HERE
```

Hàm cross_validate sau khi thực hiện xong tính toán sẽ trả về đầu ra dưới dạng một từ điển. Trong đó: có các giá trị sau cần chú ý:
- estimator: Mô hình đã được tối ưu
- train_precision/test_precision: Các thông số thể hiện performance của mô hình trên hai tập train và test

```python
sorted(scores.keys())
```

```python
print(scores['train_precision_macro'])
print(scores['test_precision_macro'])
```

```python
print(scores['train_precision_macro'].mean())
print(scores['test_precision_macro'].mean())
```

```python
print(scores['train_recall_macro'])
print(scores['test_recall_macro'])
```

```python
print(scores['train_recall_macro'].mean())
print(scores['test_recall_macro'].mean())
```

```python
scores["estimator"][0]
```

```python
# Gõ key estimator để  truy xuất ra các mô hình trong quá trình cross val
# Gõ coeff_ để truy xuất ra trọng số của mô hình
scores["estimator"][0].coef_
```
